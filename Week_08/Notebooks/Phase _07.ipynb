```python
# Feature Importance
importance = pd.Series(
    model.feature_importances_,
    index=df.drop("SalePrice", axis=1).columns
)

importance = importance.sort_values(ascending=False)

importance.head(15)
```

```python
# Feature Importance Chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

importance.head(15).plot(kind="barh")

plt.title("Top 15 Important Features")

plt.xlabel("Importance")

plt.show()
```

# Feature Importance

The feature importance analysis indicates that the following variables have the greatest impact on house price prediction:

These features contribute significantly to the prediction accuracy of the selected regression model.

# Comparison of Regression Models

The following regression models were developed and evaluated:

- Linear Regression
- Decision Tree Regressor
- Random Forest Regressor
- Gradient Boosting Regressor
- XGBoost Regressor
- LightGBM Regressor
- CatBoost Regressor

Each model was evaluated using the same training and testing datasets to ensure a fair comparison.

The evaluation metrics included:

- MAE
- MSE
- RMSE
- R² Score
- Cross Validation Score

The model with the highest R² Score and the lowest prediction error was selected as the final model.
