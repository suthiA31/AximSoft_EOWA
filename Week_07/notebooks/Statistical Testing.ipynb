# Phase 5 - Statistical Testing

## Objectives

This phase performs hypothesis testing to identify statistically significant relationships within the hotel booking dataset.

### Parametric Tests
- Independent t-Test
- Paired t-Test
- One-Way ANOVA
- Pearson Correlation

### Non-Parametric Tests
- Chi-Square Test
- Mann-Whitney U Test
- Kruskal-Wallis Test
- Spearman Correlation

Each statistical test includes:

- Null Hypothesis (H₀)
- Alternative Hypothesis (H₁)
- Assumption Validation
- Test Statistic
- p-value
- Decision
- Business Interpretation

```python
import pandas as pd
import numpy as np

from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
from scipy.stats import f_oneway
from scipy.stats import pearsonr
from scipy.stats import chi2_contingency
from scipy.stats import mannwhitneyu
from scipy.stats import kruskal
from scipy.stats import spearmanr

import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv("processed_hotel_bookings.csv")
print("H0 : Mean ADR is equal for both hotel types")
print("H1 : Mean ADR differs between hotel types")
```

```python
city = df[df["hotel"]==0]["adr"]

resort = df[df["hotel"]==1]["adr"]

stat,p = ttest_ind(city,resort)

print("Statistic :",stat)
print("P-value :",p)
```

```python
if p<0.05:
    print("Reject H0")
    print("Business: ADR differs significantly between hotel types.")
else:
    print("Fail to Reject H0")
```

```python
print("H0 : Mean weekend stay equals weekday stay")
print("H1 : Mean weekend stay differs from weekday stay")
```

```python
stat,p = ttest_rel(
    df["stays_in_weekend_nights"],
    df["stays_in_week_nights"]
)

print("Statistic :",stat)
print("P-value :",p)
```

```python
if p<0.05:
    print("Reject H0")
    print("Business: Weekend and weekday stays differ.")
else:
    print("Fail to Reject H0")
print("H0 : Mean ADR is same across customer types")
print("H1 : At least one customer type differs")    
```

```python
groups=[]

for i in df["customer_type"].unique():
    groups.append(df[df["customer_type"]==i]["adr"])

stat,p = f_oneway(*groups)

print("Statistic :",stat)
print("P-value :",p)
```

```python
if p<0.05:
    print("Reject H0")
    print("Business: Customer type affects ADR.")
else:
    print("Fail to Reject H0")
```

```python
corr,p = pearsonr(
    df["lead_time"],
    df["adr"]
)

print("Correlation :",corr)
print("P-value :",p)
```

```python
if p<0.05:
    print("Reject H0")
    print("Business: Lead time significantly influences ADR.")
else:
    print("Fail to Reject H0")
```

```python
#One-Way ANOVA

print("H0 : Mean ADR is same across customer types")
print("H1 : At least one customer type differs")

groups=[]

for i in df["customer_type"].unique():
    groups.append(df[df["customer_type"]==i]["adr"])

stat,p = f_oneway(*groups)

print("Statistic :",stat)
print("P-value :",p)

if p<0.05:
    print("Reject H0")
    print("Business: Customer type affects ADR.")
else:
    print("Fail to Reject H0")
```

```python
#Pearson Correlation
print("H0 : No linear relationship exists")
print("H1 : Linear relationship exists")

corr,p = pearsonr(
    df["lead_time"],
    df["adr"]
)

print("Correlation :",corr)
print("P-value :",p)

if p<0.05:
    print("Reject H0")
    print("Business: Lead time significantly influences ADR.")
else:
    print("Fail to Reject H0")
```

```python
#Chi-Square Test

print("H0 : Hotel type and cancellation are independent")
print("H1 : Hotel type and cancellation are associated")

table = pd.crosstab(
    df["hotel"],
    df["is_canceled"]
)

chi,p,dof,expected = chi2_contingency(table)

print("Chi-square :",chi)
print("P-value :",p)

if p<0.05:
    print("Reject H0")
    print("Business: Cancellation depends on hotel type.")
else:
    print("Fail to Reject H0")
```

```python
#Mann-Whitney U Test

print("H0 : ADR distributions are equal")
print("H1 : ADR distributions differ")

stat,p = mannwhitneyu(city,resort)

print("Statistic :",stat)
print("P-value :",p)

if p<0.05:
    print("Reject H0")
    print("Business: ADR differs significantly.")
else:
    print("Fail to Reject H0")
```

```python
#Kruskal-Wallis Test

print("H0 : Median ADR is equal")
print("H1 : At least one group differs")

stat,p = kruskal(*groups)

print("Statistic :",stat)
print("P-value :",p)

if p<0.05:
    print("Reject H0")
    print("Business: Customer groups have different ADR.")
else:
    print("Fail to Reject H0")
```

```python
#Spearman Correlation

print("H0 : No monotonic relationship")
print("H1 : Monotonic relationship exists")

corr,p = spearmanr(
    df["lead_time"],
    df["adr"]
)

print("Correlation :",corr)
print("P-value :",p)

if p<0.05:
    print("Reject H0")
    print("Business: Lead time has significant monotonic relationship with ADR.")
else:
    print("Fail to Reject H0")
```
